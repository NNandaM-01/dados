```python
import pandas as pd
import matplotlib.pyplot as plt
pd.read_excel('/content/Tabela 9514.xlsx',skiprows= 4)
população = pd.read_excel('/content/Tabela 9514.xlsx',skiprows=4)
população.head()
população = população.dropna()
novo_nome_colunas = {
      'Tabela 9514 - População residente, por sexo, idade e forma de declaração da idade': 'nome dos Estados',
      'Unnamed: 1':'totalporEstado',
      'Unnamed: 2': 'totalhomem',
      'Unnamed: 3': 'totalmulher'}
população = população.rename(columns=novo_nome_colunas)
print(população.head())

# Plotar gráfico de barras para a população por estado desejado
estados_desejados = ['São Paulo', 'Rio de Janeiro', 'Minas Gerais','Distrito Federal','Goiás','Mato Grosso']
população_filtrados = população[população['Idade - Total'].isin(estados_desejados)]

plt.figure(figsize=(10, 6))
plt.bar(população_filtrados['Idade - Total'], população_filtrados['totalporEstado'], color='skyblue')
plt.xlabel('Estados')
plt.ylabel('População')
plt.title('População por Estado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

total_homens =população['totalhomem'].sum()
total_mulheres = população['totalmulher'].sum()

print("Total de Homens:", total_homens)
print("Total de Mulheres:", total_mulheres)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(população['Idade - Total'], população['totalporEstado'], color='skyblue')
plt.xlabel('Estados')
plt.ylabel('População')
plt.title('População por Estado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
for estado in estados_desejados:
    população_estado = população_filtrados[população_filtrados['Idade - Total'] == estado]
    total_homens = população_estado['totalhomem'].iloc[0]
    total_mulheres = população_estado['totalmulher'].iloc[0]
    plt.figure(figsize=(6, 6))
    plt.pie([total_homens, total_mulheres], labels=['Homens', 'Mulheres'], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
    plt.title(f'Distribuição da População por Gênero em {estado}')
    plt.show()
```
