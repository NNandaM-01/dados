criamos um intervalo de 5 em 5 anos para  ver a frequencia de idade que mais temos.

```python

# Calcular a idade total em anos
MB['Idade total (anos)'] = MB['Idade (anos)'] + MB['Idade (meses)']

# Definir os intervalos de idade em anos
intervalos = range(27, 65, 5)  # Intervalos de 5 anos: 20-24, 25-29, 30-34, 35-39, 40-44, 45-49

# Definir os rótulos dos intervalos (5 rótulos para 6 intervalos)
rotulos = [f'{i}-{i+5} anos' for i in intervalos[:-1]]  # Remove o último intervalo

# Categorizar as idades em intervalos
MB['Intervalo'] = pd.cut(MB['Idade total (anos)'], bins=intervalos, labels=rotulos, right=False)

# Contar as ocorrências em cada intervalo
contagem_intervalos = MB['Intervalo'].value_counts().sort_index()

# Plotar o gráfico de barras
contagem_intervalos.plot(kind='bar', color='blue')

# Adicionando rótulos e título ao gráfico
plt.title('Distribuição de Idades por Intervalo de 5 Anos')
plt.xlabel('Intervalo de Idade')
plt.ylabel('Número de Pessoas')

# Ajustando as etiquetas do eixo x
plt.xticks(rotation=45)

# Exibir o gráfico
plt.show()
```

```python
color_map = {'interior': 'blue', 'capital': 'green', 'outra': 'red'}
colors = MB['Região de procedência'].map(color_map)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(MB['Idade (anos)'],MB['Salário (x sal. min.)'], c=colors, s=100, alpha=0.75, edgecolors='w')
plt.xlabel('Idade (anos)')
plt.ylabel('Salário (x sal. min.)')
plt.title('Relação entre Idade e Salário')
plt.colorbar()

legend_labels = [plt.Line2D([0], [0], marker='o', color='w', label='Interior', markerfacecolor='blue', markersize=10),
                 plt.Line2D([0], [0], marker='o', color='w', label='Capital', markerfacecolor='green', markersize=10),
                 plt.Line2D([0], [0], marker='o', color='w', label='Outra', markerfacecolor='red', markersize=10)]
plt.legend(handles=legend_labels, title='Região de Procedência', loc='upper left')

# Exibindo o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
grouped = MB.groupby(['Estado', 'N° de filhos']).size().unstack()

# Plotando o gráfico de barras agrupadas
grouped.plot(kind='bar', stacked=False, figsize=(10, 6))

# Adicionando rótulos e título
plt.title('Número de Pessoas por Estado Civil e Número de Filhos')
plt.xlabel('Estado Civil')
plt.ylabel('Número de Pessoas')
plt.xticks(rotation=0)
plt.legend(title='N° de filhos', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
bubble_sizes = MB['N° de filhos'] * 100
# Plotando o gráfico de dispersão com bolhas
plt.figure(figsize=(10, 6))
plt.scatter(MB['Grau de instrução'], MB['Idade (anos)'], s=bubble_sizes, alpha=0.7, c='green', edgecolors='w')

plt.title('Relação entre Grau de Instrução e Idade')
plt.xlabel('Grau de Instrução')
plt.ylabel('Idade')

plt.grid(True)
plt.tight_layout()
plt.show()
```
